```python
import json

import pandas as pd
from pathlib import Path
pd.set_option('max_colwidth',300)
from pprint import pprint
```

```python
f = Path('python/final/jsonl/test/python_test_0.jsonl')
df = pd.read_json(f, orient='records', lines=True)
df
```

```python
columns_short_list = ['code_tokens', 'docstring_tokens', 
                      'language', 'partition']
temp_df = df.head()[columns_short_list]
temp_df
```

```python
temp_df.partition.value_counts()
```

```python
temp_df.groupby(['partition', 'language'])['code_tokens'].count()
```

```python
temp_df['code_len'] = temp_df['code_tokens'].apply(lambda x: len(x))
temp_df
```

```python
temp_df['query_len'] = temp_df['docstring_tokens'].apply(lambda x: len(x))
temp_df
```
